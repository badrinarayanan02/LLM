Multimodal Input

```python
image_urls = [
    "https://c8.alamy.com/comp/EX6P38/colour-satellite-image-of-bangalore-india-image-taken-on-march-31-EX6P38.jpg"
]

image_documents = load_image_urls(image_urls)
gemini = GeminiMultiModal(model_name="models/gemini-1.5-flash")
```

Inference: Used gemini-1.5-flash model from google to handle multimodality.

Displaying the Image

```python
img_response = requests.get(image_urls[0])
print(image_urls[0])
img = Image.open(BytesIO(img_response.content))
plt.imshow(img)
```

Specifying prompt

```python
response = gemini.complete(
    prompt = "Identify the city and convey where are the green spaces and parks located?",
    image_documents = image_documents
)
```
